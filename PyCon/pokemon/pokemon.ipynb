```python
# all at once
stats = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']

pokemon = pokemon[
    [
        'english_name',
        'primary_type',
        'hp',
        'attack',
        'defense',
        'sp_attack',
        'sp_defense',
        'speed',
        'is_sublegendary',
        'is_legendary',
        'is_mythical',
    ]
].astype(
    {
        'primary_type': 'category',
        'is_sublegendary': 'bool',
        'is_legendary': 'bool',
        'is_mythical': 'bool',
    }
).join(
    2 * pokemon[stats] + (110, 5, 5, 5, 5, 5),
    rsuffix='_final',
).assign(
    total = pokemon[stats].sum(axis=1),
    is_legendary = pokemon[['is_sublegendary', 'is_legendary', 'is_mythical']].any(axis=1)
)
```

```python
type_colors_data = {
    "normal": "#A8A77A",
    "fire": "#EE8130",
    "water": "#6390F0",
    "electric": "#F7D02C",
    "grass": "#7AC74C",
    "ice": "#96D9D6",
    "fighting": "#C22E28",
    "poison": "#A33EA1",
    "ground": "#E2BF65",
    "flying": "#A98FF3",
    "psychic": "#F95587",
    "bug": "#A6B91A",
    "rock": "#B6A136",
    "ghost": "#735797",
    "dragon": "#6F35FC",
    "dark": "#705746",
    "steel": "#B7B7CE",
    "fairy": "#D685AD",
}
```

```python
fig, ax = plt.subplots(
    figsize=(8, 4.5),
    facecolor="whitesmoke",
    layout="constrained",
)

primary_type = pokemon['primary_type']

ax.bar(
    range(primary_type.nunique()),
    primary_type.value_counts(),
    tick_label=primary_type.cat.categories,
    color=type_colors_data.values()
)
ax.set_xticks(
    range(primary_type.nunique()),
    labels=primary_type.value_counts().index,
    rotation=90
)

ax.set_title("Distribution of Pokemon per type")
ax.set_xlabel("Primary type")
ax.set_ylabel("Count")
```

```python
fig, ax = plt.subplots(
    figsize=(8, 4.5),
    facecolor="whitesmoke",
    layout="constrained",
)

primary_type = pokemon['primary_type']
value_counts = primary_type.value_counts()
colors = [type_colors_data[color] for color in value_counts.index]

ax.bar(
    range(primary_type.nunique()),
    value_counts,
    color=colors
)
ax.set_xticks(
    range(primary_type.nunique()),
    labels=primary_type.value_counts().index,
    rotation=90
)

ax.set_title("Distribution of Pokemon per type")
ax.set_xlabel("Primary type")
ax.set_ylabel("Count")
```

```python
fig, ax = plt.subplots(
    figsize=(8, 4.5),
    facecolor="whitesmoke",
    layout="constrained",
)

primary_type = pokemon['primary_type']
value_counts = primary_type.value_counts()
colors = [type_colors_data[color] for color in value_counts.index]

ax.barh(
    range(primary_type.nunique()),
    value_counts,
    color=colors
)
ax.set_yticks(
    range(primary_type.nunique()),
    labels=primary_type.value_counts().index,
)

ax.set_title("Distribution of Pokemon per type")
ax.set_ylabel("Primary type")
ax.set_xlabel("Count")
fig.savefig('best_plot.png')
```
